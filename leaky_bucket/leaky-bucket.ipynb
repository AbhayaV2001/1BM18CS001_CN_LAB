```python
class LeakyBucket:
  def __init__(self, bucket_size, output, packets):
    self.bucket_size = bucket_size
    self.output = output
    self.packets = packets

  def congestion_control(self):
    for i in range(len(self.packets)):
      print(f"Packet No: {i} Packet Size: {packets[i]}")

      if packets[i] > self.bucket_size:
        print("\t Bucket Overflow")
      else:
        while packets[i] > self.output:
          print(f"\t {self.output} bytes sent")
          packets[i] -= self.output

        if packets[i]:
          print(f"\t Last {packets[i]} bytes sent")
        
        print("\t Bucket output successful \n")


bucket_size = int(input("Enter Bucket Size: "))
output = int(input("Enter Output Rate: "))
packets = list(map(int, input("Enter Input Packets: ").split()))

network = LeakyBucket(bucket_size, output, packets)
network.congestion_control()
```
